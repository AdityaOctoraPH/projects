```python
import numpy as np
```

```python
import matplotlib.pyplot as mlt
```

```python
import pandas as pd
```

```python
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
print(X)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
print(X_train)
```

```python
from sklearn.linear_model import LinearRegression
reggresor = LinearRegression()
reggresor.fit(X_train, y_train)
```

```python
mlt.scatter(X_train, y_train, color='red')
mlt.plot(X_train, reggresor.predict(X_train))
mlt.title('Training Set')
mlt.xlabel('Year(s) Experience')
mlt.ylabel('Salary')
mlt.show()
```

```python
mlt.scatter(X_test, y_test, color='red')
mlt.plot(X_train, reggresor.predict(X_train))
mlt.title('Test Set')
mlt.xlabel('Year(s) Experience')
mlt.ylabel('Salary')
mlt.show()
```
